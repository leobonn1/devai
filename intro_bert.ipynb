```python
from transformers import pipeline
```

```python
unmasker = pipeline(task='fill-mask', model='bert-base-uncased')
```

```python
unmasker("Hello I'm a [MASK] model.")
```

```python
from transformers import pipeline

pipe = pipeline(task='fill-mask', model='neuralmind/bert-base-portuguese-cased')
```

```python
pipe('Tinha uma [MASK] no meio do caminho.')
```

```python
pipe('Os [MASK] de declaração possuem efeito devolutivo restrito e fundamentação vinculada.')
```

```python
pipe_legal_1 = pipeline(task='fill-mask', model='raquelsilveira/legalbertpt_sc')

pipe_legal_1('Os [MASK] de declaração possuem efeito devolutivo restrito e fundamentação vinculada.')
```

```python
pipe_legal_2 = pipeline(task='fill-mask', model='dominguesm/legal-bert-base-cased-ptbr')

pipe_legal_2('Os [MASK] de declaração possuem efeito devolutivo restrito e fundamentação vinculada.')
```
